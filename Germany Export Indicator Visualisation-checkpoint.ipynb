### Further exploring GERMANY in the World Development Indicators Dataset


```python
import pandas as pd
```

```python
# Read in the World Development Indicators Database
# We can have a look at database structure. It has 5656458 rows and 6 columns. Enormous dataset!
data = pd.read_csv('C:/Users/Admin\Desktop/data science/week 5 visual/Indicators.csv')
data.shape
```

```python
data.head()
```

```python
# Display first and last rows
data.iloc[[1,2,-1,-2]]
```

```python
# as first step, filter database for Germany
germany=data['CountryName'].str.contains("Germany")

# Dataset of Germany
db_of_germany =data[germany]

# Display first 10 rows dataset for Germany
db_of_germany.head(10)
```

```python
# Display how many avaible indicators in the dataset 
len(db_of_germany['IndicatorName'].unique().tolist())
```

As we can see above , There is 879 diffirent indicators exist in the dataset

```python
# Display dataset all indicators related to Germany after 2000
# Purpose of this approach is to analyze 21th century
recent_years = db_of_germany['Year'] > 2000

# Display database of Germany for selected years
germany_recent=db_of_germany[recent_years]
germany_recent.head()
```

```python
# Let's pick up one indicator we want to analyze and list
indicator ='Export value index'

# This allows us to list selected indicator from dataset
export_of_deu=germany_recent['IndicatorName'].str.contains(indicator)

deu_export=germany_recent[export_of_deu]

# Display export value after 2000 years for Germany
deu_export
```

### Setup our data for plotting.  

Create a data frame with just the country codes and the values we want plotted.

```python
import matplotlib.pyplot as plt
# get the years
years = deu_export['Year'].values
# get the values 
exp =  deu_export['Value'].values

# Label the axes
plt.xlabel('Year')
plt.ylabel(deu_export['IndicatorName'].iloc[0])


# Labet for the plot
plt.title('Export in Germany for 21th century')
# create
plt.bar(years,exp) 
plt.show()
```

```python
# switch to a line plot
plt.plot(deu_export['Year'].values, deu_export['Value'].values)

# Label the axes
plt.xlabel('Year')
plt.ylabel(deu_export['IndicatorName'].iloc[0])

#label the figure
plt.title('Exports in Germany in 21th century')

# to make more honest, start they y axis at 0
plt.axis([2001, 2015,0,300])

plt.show()
```

Now let's create one more dataframe about imports and compare to export values

```python
# Let's pick up one indicator we want to analyze and list
indicator ='Import value index'

# This allows us to list selected indicator from dataset
import_of_deu=germany_recent['IndicatorName'].str.contains(indicator)

deu_import=germany_recent[import_of_deu]

# Display export value after 2000 years for Germany
deu_import
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig, axis = plt.subplots()
# Grid lines, Xticks, Xlabel, Ylabel

axis.yaxis.grid(True)
axis.set_title('Correlation between import and export values',fontsize=15)
axis.set_xlabel(deu_export['IndicatorName'].iloc[0],fontsize=10)
axis.set_ylabel(deu_import['IndicatorName'].iloc[0],fontsize=10)

X = deu_export['Value']
Y = deu_import['Value']

axis.scatter(X, Y)
plt.show()
```
